```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# data import 
data=pd.read_csv('5209_ML_Data.csv')
df_data_col = pd.DataFrame(data).columns
data.head()
```

```python
# data shape
print("data shape:", data.shape)

# check missing value 
data.isnull().sum()
```

```python
# data overview
data.describe()
```

```python
# heatmap - rough idea of variable relationship 
#sns.heatmap(data.corr(),annot=True,cmap='RdYlGn',linewidths=0.2) #data.corr()-->correlation matrix
#fig=plt.gcf()
#fig.set_size_inches(14, 14)
#plt.show()
```

```python
values = data.values     #df -> array
data_ = values[:-1,2:]   #data input  - remove 2 columns (SPX T+1 & date) 
target_ = values[:-1,:1].reshape(-1,1)     


#variable_ : temp variable
#check variable
print(data_.shape, target_.shape)
print(data_[:3])
print(target_[:3])
```

```python
# StandardScaler - in order to standardie data scale

from sklearn.preprocessing import LabelEncoder, StandardScaler   
x_scaler = StandardScaler()
data = x_scaler.fit_transform(data_)

y_scaler = StandardScaler()
target = y_scaler.fit_transform(target_)   

#print(pd.DataFrame(data_).head(3))
#print(pd.DataFrame(data).head(3))
#print(pd.DataFrame(target_).head(3))
#print(pd.DataFrame(target).head(3))

print("data size: ", data.shape, target.shape)
```

```python
train_num = int(0.8786 * data.shape[0])
X_train = data[:train_num, :]
X_test = data[train_num:, :]
y_train = target[:train_num, :]
y_test = target[train_num:, :]
# front 87.8% - train data / test data split, in line wiht course specification (2022M6 -M8 as test data)
# time series data -> make sure we won't use future data to forecast past 

print("data size: ", X_train.shape, X_test.shape)
print("data size: ", y_train.shape, y_test.shape)
```

```python
#from tensorflow.keras.models import Sequential
from tensorflow.keras.models import Sequential, load_model
from keras.layers import Activation
from keras.layers import Dense
from keras.layers import BatchNormalization
from tensorflow.keras import layers
```

```python
# step 1 定義model長什麼樣子

model = Sequential()                     # sequential nn model
model.add(layers.Dense(32,input_shape=(23,)))   # input_shape=(23,) 
model.add(layers.BatchNormalization())   # Batch Normalization explain  http://violin-tao.blogspot.com/2018/02/ml-batch-normalization.html
model.add(layers.Activation('relu'))     # 用ReLU as activation function 
model.add(layers.Dense(32))
model.add(layers.Activation('relu'))
model.add(layers.Dense(32))
model.add(layers.Activation('relu'))
model.add(layers.Dense(32))
model.add(layers.Activation('relu'))
model.add(layers.Dense(32))
model.add(layers.Activation('relu'))
model.add(layers.Dense(32))
model.add(layers.Dense(1))

print(model.summary())
```

```python
# define Loss function & optimizer 
from keras.optimizers import Adam
model.compile(loss='mape', optimizer='Adam')   

# optimizer='Adam' --->  DL usually uses Adam optimizer 
```

```python
# step 3: train model    :  fit network
history = model.fit(X_train, y_train, epochs=100, batch_size=100, 
                    validation_split=0.15, shuffle=True)
#  epochs=100     -> whole dataset runs 100 times 
#  batch_size=100 -> for each run, batch size = 100
#  so total iteration = (dataset size / batch size) * epoch 

history
```

```python
from matplotlib import pyplot
# plot ML history to see validation score 
# choose the model with decreasing and stable validation score
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='validation')
pyplot.title('Loss')
pyplot.legend()
pyplot.show()
```

```python
# TRAINING DATA

from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score
# make a prediction
y_pred = model.predict(X_train)

# calculate r2 & MAPE
rmse = sqrt(mean_squared_error(y_scaler.inverse_transform(y_train), y_scaler.inverse_transform(y_pred)))
r2 = r2_score(y_scaler.inverse_transform(y_train), y_scaler.inverse_transform(y_pred))
mape_value = mape(y_scaler.inverse_transform(y_train), y_scaler.inverse_transform(y_pred))

# plot y predict and y train set
pyplot.figure(figsize=(20, 6))
pyplot.plot(range(y_train[:1000].shape[0]), y_scaler.inverse_transform(y_train[:1000]),'b-')  # b-  = Blue line
pyplot.plot(range(y_pred[:1000].shape[0]), y_scaler.inverse_transform(y_pred[:1000]),'r.')    # r.  = Red Dot
# inverse transform y_train, y_pred to origional scale, as they were standardized earlier
pyplot.title('Training Data Set-  R2: %0.3f  MAPE: %0.3f'%(r2, mape_value))
pyplot.xlabel('Time')
pyplot.ylabel('SPX')
pyplot.show()
```

```python
# TESTING DATA
from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score
# make a prediction
y_pred = model.predict(X_test)

# calculate r2 & MAPE
rmse = sqrt(mean_squared_error(y_scaler.inverse_transform(y_test), y_scaler.inverse_transform(y_pred)))
r2 = r2_score(y_scaler.inverse_transform(y_test), y_scaler.inverse_transform(y_pred))
mape_value = mape(y_scaler.inverse_transform(y_test), y_scaler.inverse_transform(y_pred))

# plot y predict and y train set
pyplot.figure(figsize=(20, 6))
pyplot.plot(range(y_test[:1000].shape[0]), y_scaler.inverse_transform(y_test[:1000]),'b-')  # b-  = 藍線
pyplot.plot(range(y_pred[:1000].shape[0]), y_scaler.inverse_transform(y_pred[:1000]),'r.')    # r.  = 紅點
# inverse transform y_test, y_pred to origional scale, as they were standardized earlier
pyplot.title('Testing Data-  R2: %0.3f  MAPE: %0.3f'%(r2, mape_value))
pyplot.xlabel('Time')
pyplot.ylabel('SPX')
#pyplot.show()
```
